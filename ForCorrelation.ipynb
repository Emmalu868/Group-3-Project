## For airline stock price pct_change dataset

```python
air_pct_change = pd.read_csv('./Resources/airline_monthly_return.csv', parse_dates=True)
air_pct_change['Date']= df.Date.tolist()[1:]
```

```python
air_pct_change.rename(columns={"Air Canada":'Canada',
                    "Air France in CAD": 'France',
                    "UK TUI Airways in CAD":'UK',
                   "Germany LHA in CAD": 'Germany',
                   "Japan All Nippon Airways in CAD" : 'Japan',
                   "American Airlines in CAD":'USA'},
          inplace=True, errors='raise')
```

```python
air_pct_change['Year'] = air_pct_change['Date'].dt.year
```

```python

air_pct_change
```

```python
air_pct_change.to_csv('./Resources/corr_airPctChange.csv', sep=',', encoding='utf-8', index=False)
```

```python
def melt_dataset(df):
    data = df.drop(['Year'], axis = 1, inplace = False) 
    data = data.set_index('Date').T.rename_axis('regions').rename_axis(columns=None).reset_index()
    time_lines = data.columns.to_list()[1:]
    df_melt = pd.melt(data,id_vars=['regions'],
        value_vars= time_lines,
        var_name='day',value_name='price', ignore_index=False)

    df_melt.sort_values(["day","regions"], inplace=True)
    df_melt['Year'] = df_melt['day'].dt.year
    return df_melt
```

```python
data = melt_dataset(air_pct_change)
data
```

```python
data.to_csv('./Resources/melt_airPctChange.csv', sep=',', encoding='utf-8', index=False)
```
