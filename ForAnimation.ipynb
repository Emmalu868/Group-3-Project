```python
import pandas as pd
import numpy as np
```

Data sources: https://data.ontario.ca/dataset/gasoline-report-international-gasoline-prices

```python
df = pd.read_csv('./Resources/internationalpumppricesall.csv', parse_dates=True)
```

```python
df = df[df['Tax Status']=='Total']
```

```python
df.rename(columns={"UK/Royaume-Uni":'UK',
                   "Germany/Allemagne": 'Germany',
                   "Japan/Japon" : 'Japan',
                   "USA/États-Unis d'Amérique":'USA'},
          inplace=True, errors='raise')
```

```python
df.drop(['Tax Status', 'Situation fiscale'], axis = 1, inplace = True) 
```

```python
df['Date']= pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
df
```

```python
df = df.set_index('Date').T.rename_axis('regions').rename_axis(columns=None).reset_index()
df
```

```python
df.to_csv('./Resources/for_animation.csv', sep=',', encoding='utf-8', index=False)
```

```python
time_lines = df.columns.to_list()[1:]
```

```python
df1 = pd.melt(df,id_vars=['regions'],
    value_vars= time_lines,
    var_name='day',value_name='price', ignore_index=False)

df1.sort_values(["day","regions"], inplace=True)
```

```python
df1['Year'] = df1['day'].dt.year
```

```python
df1.head(18)
```

```python
df1.to_csv('./Resources/for_animation_year.csv', sep=',', encoding='utf-8', index=False)
```

```python
dff = df1.copy()
```

```python
dff['iso_alpha'] = dff['regions']
dff
```

```python
dff['iso_alpha'] = dff['iso_alpha'].str.replace('Canada','CAN')
```

```python
dff['iso_alpha'] = dff['iso_alpha'].str.replace('France','FRA')
dff['iso_alpha'] = dff['iso_alpha'].str.replace('Germany','DEU')
dff['iso_alpha'] = dff['iso_alpha'].str.replace('Japan','JPN')
dff['iso_alpha'] = dff['iso_alpha'].str.replace('UK','GBR')
#dff['iso_alpha'] = dff['iso_alpha'].str.replace('USA','USA')
dff
```
